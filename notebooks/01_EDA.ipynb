```python
df_items = pd.read_csv(r'../data/raw/olist_order_items_dataset.csv')
df_items.head()
```

```python
df_items['cost'] = df_items['price']+df_items['freight_value']
```

```python
print("length of dataframe: ",len(df_items))
for column in df_items.columns:
    print(f"total unique values in {column}: ",df_items[column].nunique())
```

```python
#checking overlap between order ids in orders vs items dataset
len(set(df_orders['order_id'].unique())-set(df_items['order_id'].unique())),\
len(set(df_items['order_id'].unique())-set(df_orders['order_id'].unique()))
```

```python
#Trying to see if there's any pattern in this missing 775 order_id
missing_orders = list(set(df_orders['order_id'].unique())-set(df_items['order_id'].unique()))
print("orders with no items: ",len(missing_orders))
status_of_missing = df_orders[df_orders['order_id'].isin(missing_orders)]['order_status'].value_counts()
print("delivery status of breakdown of missing orders: \n",status_of_missing)
print("Missing items in delivered orders: ",len(df_orders[(df_orders['order_id'].isin(missing_orders))&\
    (df_orders['order_status']=='delivered')]))
```

```python
df_items.describe()
```

```python
#joining order_items with orders dataset to get delivery status
print(len(df_orders),len(df_items))
df_items_orders = df_items.merge(df_orders,on='order_id',how='left')
print(len(df_items_orders))
df_items_orders.head()
```

```python

order_summary = df_items_orders.groupby(['order_id','order_status']).agg(products_per_order=("product_id","count"),
                                               unique_products_per_order = ("product_id","nunique"),
                                               order_price = ("price","sum"),
                                               order_freight = ("freight_value","sum"),
                                               order_cost = ('cost',"sum")).reset_index()
order_summary[['products_per_order','unique_products_per_order','order_price','order_freight','order_cost']].mean()
```

```python
status_summary = order_summary.groupby('order_status')\
    [['products_per_order','unique_products_per_order','order_price','order_freight','order_cost']].mean()
status_summary
```

```python
df_delivered = df_items_orders[df_items_orders['order_status'].isin(['delivered'])]

df_product_stats = df_delivered.groupby("product_id").agg(total_quantity=("product_id","count"),
                                                     total_price = ("price","sum"),
                                                     total_freight = ("freight_value","sum"),
                                                     total_cost = ("cost","sum"))
#Top Products contributing to 80% of orders
df_top_80_orders_qty = df_product_stats.sort_values(by='total_quantity',ascending=False)
df_top_80_orders_qty['cum_qty'] = df_top_80_orders_qty['total_quantity'].cumsum()
df_top_80_orders_qty['cum_pct'] = df_top_80_orders_qty['cum_qty']/df_top_80_orders_qty['total_quantity'].sum()
df_top_80_orders_qty = df_top_80_orders_qty[df_top_80_orders_qty['cum_pct']<=0.8]
print("Total number of products contributing to 80% of orders quantity: ",len(df_top_80_orders_qty),"out of: ",len(df_product_stats),
      "ratio: ",len(df_top_80_orders_qty)/len(df_product_stats))
df_top_80_orders_qty.head()
```

```python
#Top Products contributing to 80% of total cost
df_top_80_orders_cost = df_product_stats.sort_values(by='total_cost',ascending=False)
df_top_80_orders_cost['cum_cost'] = df_top_80_orders_cost['total_cost'].cumsum()
df_top_80_orders_cost['cum_pct'] = df_top_80_orders_cost['cum_cost']/df_top_80_orders_cost['total_cost'].sum()
df_top_80_orders_cost = df_top_80_orders_cost[df_top_80_orders_cost['cum_pct']<=0.8]
print("Total number of products contributing to 80% of cost: ",len(df_top_80_orders_cost),"out of: ",len(df_product_stats),
      "ratio: ",len(df_top_80_orders_cost)/len(df_product_stats))
df_top_80_orders_cost.head()
```

```python
print(len(df_customers),len(df_delivered))
df_delivered = df_delivered.merge(df_customers,on='customer_id',how='left')
print(len(df_delivered))
```

```python
df_delivered.head()
```

```python
df_unique_customer_order_pair = df_delivered[['customer_unique_id','order_id']].drop_duplicates()
orders_per_customer = df_unique_customer_order_pair.groupby('customer_unique_id')['order_id'].count()
cust_orders = orders_per_customer.sort_values(ascending=False)
(cust_orders==1).mean()
```

```python
df_unique_customer_order_pair = df_delivered[['customer_unique_id','order_id']].drop_duplicates()
orders_per_customer = df_unique_customer_order_pair.groupby('customer_unique_id')['order_id'].count()
cust_orders = orders_per_customer.sort_values(ascending=False)
print("customers with exactly one order: ",(cust_orders==1).sum(),100*(cust_orders==1).mean(),"%")
cust_share = cust_orders / cust_orders.sum()
cust_cum_share = cust_share.cumsum()
print("total customers with contributing to 80% of orders: ",(cust_cum_share<=0.8).sum(),100*(cust_cum_share<=0.8).mean(),"%")
print("total orders: ",(cust_cum_share).count())
```

```python
delivered_order_ids = set(df_delivered['order_id'].unique())
reviewed_order_ids = set(df_reviews['order_id'].unique())
mutual_reviewed_order_ids = delivered_order_ids & reviewed_order_ids
print("total orders with reviews: ",len(mutual_reviewed_order_ids),100*len(mutual_reviewed_order_ids)/len(delivered_order_ids),"%")
```
