# Order items dataset

```python
df_items = pd.read_csv(r'../data/raw/olist_order_items_dataset.csv')
df_items.head()
```

```python
df_items.info()
```

```python
df_items['shipping_limit_date'] = pd.to_datetime(df_items['shipping_limit_date'])
df_items['shipping_limit_date_only'] = df_items['shipping_limit_date'].dt.normalize()
min_date = df_items['shipping_limit_date_only'].min()
max_date = df_items['shipping_limit_date_only'].max()
gap = max_date-min_date
print("min_date: ",min_date)
print("max_date: ",max_date)
print("gap: ",gap)
```

```python
df_items['cost'] = df_items['price']+df_items['freight_value']
```

```python
print("length of dataframe: ",len(df_items))
for column in df_items.columns:
    print(f"total unique values in {column}: ",df_items[column].nunique())
```

###### Only 555 out of 1298 days data is present

```python
df_items.describe()
```

```python
#average order stats
df_avg_order = df_items.groupby('order_id').agg(num_products=("product_id","count"),
                                               unique_products = ("product_id","nunique"),
                                               avg_price = ("price","mean"),
                                               avg_freight = ("freight_value","mean"),
                                               avg_cost = ('cost',"mean")).mean()
df_avg_order
```

```python
df_product_stats = df_items.groupby("product_id").agg(total_quantity=("product_id","count"),
                                                     total_price = ("price","sum"),
                                                     total_freight = ("freight_value","sum"),
                                                     total_cost = ("cost","sum"))
#Top Products contributing to 80% of orders
df_top_80_orders_qty = df_product_stats.sort_values(by='total_quantity',ascending=False)
df_top_80_orders_qty['cum_qty'] = df_top_80_orders_qty['total_quantity'].cumsum()
df_top_80_orders_qty['cum_pct'] = df_top_80_orders_qty['cum_qty']/df_top_80_orders_qty['total_quantity'].sum()
df_top_80_orders_qty = df_top_80_orders_qty[df_top_80_orders_qty['cum_pct']<=0.8]
print("Total number of products contributing to 80% of orders quantity: ",len(df_top_80_orders_qty))
df_top_80_orders_qty.head()
```

```python
#Top Products contributing to 80% of total cost
df_top_80_orders_cost = df_product_stats.sort_values(by='total_cost',ascending=False)
df_top_80_orders_cost['cum_cost'] = df_top_80_orders_cost['total_cost'].cumsum()
df_top_80_orders_cost['cum_pct'] = df_top_80_orders_cost['cum_cost']/df_top_80_orders_cost['total_cost'].sum()
df_top_80_orders_cost = df_top_80_orders_cost[df_top_80_orders_cost['cum_pct']<=0.8]
print("Total number of products contributing to 80% of cost: ",len(df_top_80_orders_cost))
df_top_80_orders_cost.head()
```

```python
#orders shipped distribution
orders_per_day = df_items.groupby("shipping_limit_date_only")["order_id"].nunique()

orders_per_day.plot()
plt.xlabel("Date")
plt.ylabel("Number of Orders")
plt.title("Orders per Day")
plt.show()
```

```python
#Orders distribution but at monthly level
orders_per_month = df_items.groupby(df_items["shipping_limit_date_only"].dt.to_period("M"))["order_id"].nunique()
orders_per_month.plot()
plt.xlabel("Month - Year")
plt.ylabel("Number of Orders")
plt.title("Orders per Month")
plt.show()
```
